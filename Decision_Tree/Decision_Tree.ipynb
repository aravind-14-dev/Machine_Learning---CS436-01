```python
def isDataPure(data):
    
    label = data[:,-1]
    unique = np.unique(label)
    
    if len(unique) !=1:
        return False
    else:
        return True
```

```python
def mostFreq(data):
    
    label = data[:,-1]
    unique,counts = np.unique(label,return_counts=True)
    
    ind = counts.argmax()
    freq = unique[ind]
    
    return freq
```

```python
def entropy(data):
    
    label = data[:,-1]
    _,counts = np.unique(label,return_counts = True)
    prob = counts/counts.sum()
    entropy = sum(prob * -np.log2(prob))
    
    return entropy
```

```python
def attrEntropy(data_below,data_above):
    
    n = len(data_below) + len(data_above)
    p1 = len(data_below) / n
    p2 = len(data_above) / n
    
    attr_entropy = (p1*entropy(data_below) + p2*entropy(data_above))
    
    return attr_entropy
```

```python
def possibleAttrs(data):
    
    potential_attrs = {}
    _,cols = data.shape
    for i in range(cols-1):
        val = data[:,i]
        unique = np.unique(val)
        
        if len(unique) > 1:
            potential_attrs[i] = unique
    
    return potential_attrs
```

```python
def split_set(data,best_attr,bestattr_value):
    
    values = data[:,best_attr]
    
    arr1 = data[values == bestattr_value]
    arr2 = data[values != bestattr_value]
    
    return arr1, arr2
```

```python
def divAttr(data, attrs):
    
    nodeInformation = len(data)*entropy(data)
    info_gain = -math.inf
    
    for i in attrs:
        for j in attrs[i]:
            data_below,data_above = split_set(data, best_attr = i, bestattr_value = j)
            attrEnt = attrEntropy(data_below, data_above)
            
            if ((nodeInformation - attrEnt) > info_gain):
                info_gain = nodeInformation - attrEnt
                best_attr = i
                best_value = j
                
    return best_attr, best_value
```

```python
def decision_tree(df,c=0):
    
    
         if c != 0:
            data = df
        
         else:
            global attr
            attr = df.columns
            data = df.values          
    

         if(isDataPure(data)):
            classification = mostFreq(data)
        
            return classification

         else:    
            c += 1

            potential_splits = possibleAttrs(data)
            split_column, split_value = divAttr(data, potential_splits)
            arr1, arr2 = split_set(data, split_column, split_value)
        
        
            feature_name = attr[split_column]
        
            node = "{} = {}".format(feature_name, split_value)
        
            sub_tree = {node: []}
        
            left = decision_tree(arr1, c)
            right = decision_tree(arr2, c)
        
            if left == right:
                sub_tree = right
            else:
                sub_tree[node].append(left)
                sub_tree[node].append(right)
        
            return sub_tree
```

```python
def evaluate_test(test, tree):
    node = list(tree.keys())[0]
    attr_name, operator, value = node.split(" ")

    if str(test[attr_name]) == value:
        answer = tree[node][0]
    else:
        answer = tree[node][1]

    if not isinstance(answer, dict):
        return answer
    else:
        sub_tree = answer
        return evaluate_test(test, sub_tree)    
```

```python
def accuracy(df,tree):
    
    df["prediction"] = df.apply(evaluate_test, args=(tree,), axis =1)
    df["correct_prediction"] = df["prediction"] == df["Class"]
    accuracy = df["correct_prediction"].mean()
    
    return accuracy
```
