```python
import os 
import pandas as pd
from collections import Counter
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import *
stemmer = PorterStemmer()
```

```python
def stem(data):
    
    new = {}
    for k,v in data.items():
        new[k] = stemmer.stem(k)
    for old,new in new.items():
        data[new] = data.pop(old) 
        
    return data
```

```python
def train():
    data = {}
    total_spam = 0
    total_ham = 0
    nof_spam = 0
    nof_ham = 0
    
    for filename in os.listdir(".\ham_train"):
        file = open('.\ham_train' + '\\' + filename, errors='ignore')
        wordcount = Counter(file.read().split())
        #print(wordcount.keys())
        for item in wordcount.items():
            #print(item)
            if item[0] in data:
                data[item[0]][1] += item[1]
            else:
                data[item[0]] = [0,item[1]]
            #data = data + ['ham']
            total_ham += item[1]
        nof_ham += 1

    for filename in os.listdir('.\spam_train'):
        file = open('.\spam_train'+'\\' +filename,errors='ignore')
        wordcount = Counter(file.read().split())
        for item in wordcount.items():
            if item[0] in data:
                 data[item[0]][0] += item[1]
            else:
                data[item[0]] = [item[1],0]
            total_spam += item[1]
        nof_spam += 1
    #print(nof_spam)
    #print(data)
    data = stem(data)
    #data_new,ps,ph = condProb(data,total_spam,total_ham,nof_spam,nof_ham)
    #print(data_new)
    
        
    
    return data
```

```python
def rem_stop(wordcount):
    
    l = []
    stop_words = set(stopwords.words('english'))
    for k in wordcount:
        #print(k)
        if k in stop_words:
            l.append(k)
    #print(l)
    for k in l:
        del wordcount[k]
    return wordcount
```

```python
def isClass(filename):
    
    #for filename in os.listdir('.\Training'):
    output = ""
    if 'spam' in filename:
        output = 'spam'
        #return output
    if 'ham' in filename:
        output = 'ham'
        #return output
    return output
```

```python
def find(instance,weights):
    #d1,ps,ph = train()
    
    
    weight_sum = weights['weight_zero']
    
    file = open('.\Testing'+'\\' +instance,errors='ignore')
    wordcount = Counter(file.read().split())
    #wordcount = rem_stop(wordcount)
    for f in wordcount.keys():
        #print(f)
        if f not in weights:
            weights[f] = 0.0
        weight_sum += weights[f]*wordcount[f]
        #print(weight_sum)
    if weight_sum>0:
        return 1
    else:
        return 0
```

```python
def learnweights(learning_const,n):
    
    ni = n
    weights = {'weight_zero':1}
    d1 = train()
    
    
    """for filename in os.listdir('.\spam_train'):
        file = open('.\spam_train'+'\\' +filename,errors='ignore')
        wordcount = Counter(file.read().split())"""
    #print(wordcount)
    for i in d1.keys():
            weights[i] = 0.0
    #print(weights)
    for i in range(0,ni):
        
        for filename in os.listdir('.\Training'):
            k = 'spam'
            file = open('.\Training'+'\\' +filename,errors='ignore')
            wordcount = Counter(file.read().split())
            weight_sum = weights['weight_zero']
            #wordcount = rem_stop(wordcount)
        #print(weight_sum)
            for f in wordcount.keys():
            #print(weights[f])
                if f not in weights:
                    weights[f] = 0.0
                weight_sum += weights[f]*wordcount[f]
            #print(weight_sum)
            perceptron_output = 0.0
            if weight_sum>0:
                perceptron_output = 1.0
            target_value = 0.0
            
            if isClass(filename) == 'spam':
                target_value = 1.0
            for w in wordcount.keys():
                weights[w] += float(learning_const)*float((target_value - perceptron_output))*float(wordcount[w])
                #print(weights)
    
    corr_guess = 0
    count = 0.0
    for filename in os.listdir('.\Testing'):
        guess = find(filename,weights)
        if guess == 1 :
            if isClass(filename) == 'spam':
                corr_guess += 1
        if guess == 0:
            if isClass(filename) == 'ham':
                corr_guess += 1
        count += 1
    #print(count)
    accuracy = float(corr_guess)/float(count)
    print(accuracy)
```

```python
def main():
    NoI = 5
    for i in [float(j)/100 for j in range(5,105,5)]:
        learnweights(i,NoI)
        NoI += 5
```

```python
if __name__ == "__main__":
    main()
```
